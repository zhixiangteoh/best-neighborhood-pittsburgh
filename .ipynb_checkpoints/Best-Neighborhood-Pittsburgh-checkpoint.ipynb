## Bus stop density by neighborhood

Let's dive into the data analysis! First, we need to load the data, and we will do so by first **streaming** the relevant data from the WPRDC website. Let's take a look at both the data *dictionary* (which gives us useful information about the dataset fields) and the dataset.

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# dictionary
dictTS  = pd.read_csv("https://data.wprdc.org/datastore/dump/0a26e18e-f78b-43d2-8c75-2498ae255ecf")

dictTS
```

```python
mapTS = gpd.read_file("transit_stops/PAAC_Stops_1909.shp") # read in the shapefile

# recreate 'geometry' column with Latitude and Longitude data to match geometry formatting of neighborhoods shapefile
mapTS = gpd.GeoDataFrame(mapTS, geometry=gpd.points_from_xy(mapTS.Longitude, mapTS.Latitude))

mapTS.head()
```

```python
mapTS[['Stop_name', 'Latitude','Longitude','geometry']]
```

```python
# filter out non-"Bus Stops"
mapTS = mapTS[mapTS["Stop_type"] == "Bus Stop"]

plt.style.use('dark_background')
mapTS.plot(figsize=(12,8)) # plot the geopandas dataframe
```

```python
# load neighborhoods location data
neighborhoods = gpd.read_file("neighborhoods/Neighborhoods_.shp") # read in the shapefile

neighborhoods.plot(column='area', 
                   cmap='OrRd', 
                   edgecolor='orange', 
                   legend=True,
                   legend_kwds={'label':'Neighborhood Size'},
                   figsize=(12,8)) # plot the geopandas dataframe
```

```python
neighborhoods[['hood', 'area', 'perimeter', 'geometry']]
```

We want to classify the bus stops into their respective neighborhoods. Thankfully, we can do this easily using the **geopandas** library's *spatial join "sjoin"* method with the *\"contains*\" operation. This means we are checking to see which bus stops each neighborhood "contains". Looking at the new *joint* dataframe we can see that each bus stop is now associated with a **hood**, which is exactly what we want.

```python
# TSgeodf = gpd.GeoDataFrame.from_file("transit_stops/PAAC_Stops_1909.shp")
mapTS.crs = "epsg:4326"
# hoods = gpd.GeoDataFrame.from_file("neighborhoods/Neighborhoods_.shp")
neighborhoods.crs = "epsg:4326"

hoods_TS = gpd.sjoin(neighborhoods, mapTS, op="contains")

# print(TSwithHoods.columns)
hoods_TS = hoods_TS[['hood','Stop_name','geometry']]
hoods_TS
```

Now, our goal is to count the number of entries for each unique neighborhood (row index). Hence we use the *groupby* method and reset the indices. Then we **merge** our new dataframe with our *Neighborhoods* dataframe, and trim it to get the Area column, then we are ready to analyze the bus stop density per neighborhood!

```python
# group bus stops by neighborhood and create new dataframe of neighborhoods and bus stop counts
grouped_hoods_TS = hoods_TS.groupby("hood").size()
hoods_TScounts = grouped_hoods_TS.to_frame().reset_index()
hoods_TScounts.columns = ['Neighborhood','Stop_count']

hoods_TScounts = hoods_TScounts.merge(neighborhoods, how="inner", left_on="Neighborhood", right_on="hood")
hoods_TScounts.rename(columns = {'area':'Area'}, inplace = True)
hoods_TScounts = hoods_TScounts[['Neighborhood','Stop_count','Area','geometry']]
hoods_TScounts.head(10)
```

Before we dive into calculating the bus stop density per neighborhood, let's figure out what *units* the areas are in, since there is no comprehensive data dictionary on the WPRDC website for the Neighborhoods data with this information. To determine the units of representation I calculated the total of all areas which summed up to **1,423,950,712** and converted this number from different area units to square miles to compare it with Pittsburgh's total land area of **58.34 square miles**, and ascertained that it is in **square feet**. Hence I used the conversion factor of **2.788 * 10\*\*7** to convert the area to square miles.

Note: **ONLY RUN THE BELOW CELL ONCE.**

```python
# calculated total of all areas to guess and determine what units the areas are in
# hoods_TScounts['Area'].sum()

# convert area column to square miles
hoods_TScounts['Area'] = hoods_TScounts['Area'] / (2.788*10**7)

# create bus stop density (per square mile) column
hoods_TScounts['Stop_density'] = hoods_TScounts['Stop_count'] / hoods_TScounts['Area']
```

```python
hoods_TSdensity = hoods_TScounts.sort_values(by=['Stop_density'], ascending=False)
hoods_TSdensity
```

ONE issue. Perry North's area is *NaN* so its density is *NaN*. Let's fill in Perry North's area data obtained from a quick web search. :)

```python
hoods_TSdensity.loc[hoods_TSdensity['Neighborhood'] == 'Perry North', 'Area'] = 1.212
hoods_TSdensity.loc[hoods_TSdensity['Neighborhood'] == 'Perry North', 'Stop_density'] = hoods_TSdensity.loc[hoods_TSdensity['Neighborhood'] == 'Perry North', 'Stop_count'] / hoods_TSdensity.loc[hoods_TSdensity['Neighborhood'] == 'Perry North', 'Area']

# re-sort
hoods_TSdensity = hoods_TSdensity.sort_values(by=['Stop_density'], ascending=False)
hoods_TSdensity
```

Sweet! Now that that's cleared and out of the way, we can easily see which neighborhoods are in the lead in this aspect of stop density. Seems like Lincoln-Lemington-Belmar clearly leads the pack in this one. To see just how much the disparity between **Lincoln-Lemington-Belmar** and the rest of the neighborhoods is, as well as how *each neighborhood compares with the rest*, let's convert our density dataframe to a *GeoDataFrame* and plot it on a <span style="color:purple"><b>colorful</b></span> map by density!

```python
# convert to geodataframe
density = gpd.GeoDataFrame(hoods_TSdensity, geometry='geometry')

print(density[['Neighborhood','Stop_density']])

density.plot(column='Stop_density', 
             cmap='BuPu', 
             edgecolor='lightblue', 
             legend=True,
             legend_kwds={'label':'Bus Stop Density'},
             figsize=(12,8)
            )# plot the geopandas dataframe
```

So, there's that. <span style="color:#400a43"><b>Lincoln-Lemington-Belmar</b></span> and <span style="color:#8b8abb"><b>Marshall-Shadeland</b></span> are clearly in the lead for this one. They *passed* the preliminary round. Now let's examine more exciting dataâ€”on actual **ridership**. After all,

## Bus stop usage by neighborhood

```python
# load transit stop usage data
usageTS = pd.read_csv("https://data.wprdc.org/datastore/dump/3115c0b9-b48a-49aa-8e39-fd318eb62c04")
# usageTS = gpd.GeoDataFrame(usageTS, geometry=gpd.points_from_xy(usageTS.LONGITUDE, usageTS.LATITUDE))

# display usageTS columns for reference
print(usageTS.columns)

# filter to only show data for bus stops
usageTS = usageTS[usageTS["STOP_TYPE"] == "Bus"]
usageTS = usageTS[['STOP_NAME','FY19_AVG_ON','FY19_AVG_OFF','FY19_AVG_TOTAL','FY19_RANK']]
# change column names
usageTS.columns = ['Stop_name','FY19_avg_on','FY19_avg_off','FY19_avg_total','FY19_rank']
usageTS
```

Now, since we have already matched each bus stop to their neighborhood we will take advantage of this and *merge* our existing data sets instead of going through the whole *sjoin* process again.

```python
hoods_TSusage = usageTS.merge(hoods_TS, how="inner", on="Stop_name")

hoods_TSusage.rename(columns = {'hood':'Neighborhood'}, inplace = True)
hoods_TSusage[['Neighborhood','Stop_name','FY19_avg_on','FY19_avg_off','FY19_avg_total','FY19_rank','geometry']]
```

Instead of counting the number of bus stops each neighborhood has, it makes much more sense here to <i>tabulate the ranks of all the bus stops in each neighborhood and then get the **average rankings**</i>. Then we aim to rank the neighborhoods based on their average bus stop usage rankings.

```python
# group bus stops by neighborhood and create new dataframe of neighborhoods and TS usage data
grouped_hoods_TSusage = hoods_TSusage.groupby("Neighborhood")["FY19_rank"].mean().to_frame()

usage = grouped_hoods_TSusage.sort_values(by=['FY19_rank'])
usage.rename(columns = {'FY19_rank':'2019_average_usage'}, inplace = True)
usage
```

...more color? More visuals? Coming right up!

```python
usage_gdf = usage.merge(hoods_TSusage[['Neighborhood','geometry']],on="Neighborhood", how="left")
usage_gdf= gpd.GeoDataFrame(usage_gdf, geometry='geometry')

# hax to invert rankings to display on map
usage_gdf[['2019_average_usage']] = 1 / usage_gdf[['2019_average_usage']]
```

```python
usage_gdf.plot(column='2019_average_usage', 
           cmap='GnBu', 
           edgecolor='lightblue', 
           legend=True,
           legend_kwds={'label':'2019 Average Usage'},
           figsize=(12,8)
          )# plot the geopandas dataframe
```

Aaaaand... results are in! It seems like <span style="color:#19417d"><b>North Oakland</b></span> wins this one! If you observe it more closely, it makes sense that <span style="color:#245a97"><b>Central Business District</b></span> is one of the top three. <span style="color:#19417d"><b>North Oakland</b></span> and <span style="color:#225793"><b>Central Oakland</b></span> also make sense because they are the busier school districts. After all, Pittsburgh **is** a *school city*.

## Average bus ridership by neighborhood

Let's load the **usage** dataset again. We're going to need it as firepower for our last stretch. In particular, we are interested in the "ROUTE" column.

```python
usage_data = pd.read_csv("https://data.wprdc.org/datastore/dump/3115c0b9-b48a-49aa-8e39-fd318eb62c04")
usage_data
```

Now, our *main* firepower:

```python
# load average ridership data
ridershipTS = pd.read_csv("https://data.wprdc.org/datastore/dump/12bb84ed-397e-435c-8d1b-8ce543108698")

# display usageTS columns for reference
print(ridershipTS.columns)

# filter to only show data for bus stops
ridershipTS = ridershipTS[ridershipTS["mode"] == "Bus"]
ridershipTS
```

This dataset describes the **average ridership per route** in Pittsburgh. Ultimately, we want to describe the average ridership **per neighborhood**. How do we achieve this? Read on to find out! 

So first of, we want to match neighborhoods to routes. How? We first match **neighborhoods** to **bus stops**! Luckily for you (and me), we have already done that in the first section! *(Sorry if you're here before looking at the first section, please scroll up (or CTRL+F/CMD+F) to the loading and manipulation of hoods_TS)* As a quick refresher let's take a peek at the *hoods_TS* and *usage_data* dataframes that we are about to **merge**.

```python
hoods_TS.head()
```

```python
usage_data.head()
```

Let's *left* merge the two on the bus stops (i.e. "**Stop_name**"). Here, we *left* merge so that we only keep the *hoods_TS* columns that we are more interested in.

```python
hoods_routes = hoods_TS.merge(usage_data[['STOP_NAME','ROUTE']],left_on="Stop_name", right_on="STOP_NAME")
hoods_routes.rename(columns = {'hood':'Neighborhood', 'ROUTE':'route'}, inplace = True)
hoods_routes = hoods_routes.drop(columns=['STOP_NAME'])

hoods_routes
```

Sweet! Now we have a dataframe that matches each neighborhood to the bus routes it has. Next we want to *merge* this neighborhood-routes dataframe with the *ridershipTS* dataframe that we loaded in earlier to examine the **avg_riders** data column for each neighborhood.

```python
hoods_routeriders = hoods_routes.merge(ridershipTS[['route','year_month','day_type','avg_riders','day_count']],on="route")

hoods_routeriders = hoods_routeriders.sort_values(by=["avg_riders"], ascending=False)
hoods_routeriders
```

Then once again we *groupby* "Neighborhood" and take the average of the **daily average ridership per neighborhood**. Take note that here, I have simplified the problem by taking the daily average ridership across both *weekdays* and *weekends* treating them as equal; in effect I am saying that my ridership heuristic includes weekend ridership. And this is a perfectly sound heuristic. Ultimately it just depends on what you value moreâ€”since I chose the heuristic I have decided that weekend ridership is important in my determination of best neighborhood. *(I mean, if people don't even visit or at least pass by your neighborhood on weekends when they're the freest, how sad, really, is your neighborhood? JK.)*

```python
# group bus stops by neighborhood and create new dataframe of neighborhoods and route ridership
grouped_hoods_ridership = hoods_routeriders.groupby("Neighborhood")["avg_riders"].mean().to_frame()

ridership = grouped_hoods_ridership.sort_values(by=['avg_riders'], ascending=False)
ridership.rename(columns = {'avg_riders':'Average_daily_riders'}, inplace = True)
ridership
```

## Conclusion

Let's bring back the *scoring table* from the beginning to refresh our memories.

| Neighborhood      | Bus Stop Density  | Bus Stop Usage | Average Ridership | 
| ----------------- | ----------------- | -------------- | ----------------- |
| Westwood          |                   |                |                   |
| 31st ward         |                   |                |                   |
| Arlington         |                   |                |                   |

We need to come up with **weights** to assign to each heuristic or *category*, before we tabulate the rankings of neighborhoods in each category and finally decide who comes out top.

.

.

.

So after *loooong* deliberation and *caaaareful* consideration, I have <del><i>entirely arbitrarily</i></del> come up with the following weighted heuristics:
- Bus Stop Density **30%**
- Bus Stop Usage **30%**
- Average Ridership **40%**

We first look at the individual dataframes from each heuristic, and rank them individually.

```python
density['Bus Stop Density'] = range(1, len(density)+1)
density.drop(columns=['Stop_count','Area','geometry'])
```

```python
usage['Bus Stop Usage'] = range(1, len(usage) + 1)
usage
```

```python
ridership['Average Ridership'] = range(1, len(ridership) + 1)
ridership
```

```python
# create new dataframe as in scoring table, using density as basis
scores = density[['Neighborhood','Bus Stop Density']]
# merge with usage and ridership data
scores = pd.merge(scores, usage.drop(columns=['2019_average_usage']), on='Neighborhood')
scores = pd.merge(scores, ridership.drop(columns=['Average_daily_riders']), on='Neighborhood')

scores
```

**Weigh** results from the individual heuristics.

```python
weights = [0.3, 0.3, 0.4]
scores['Density weighted'] = scores['Bus Stop Density'] * weights[0]
scores['Usage weighted'] = scores['Bus Stop Usage'] * weights[1]
scores['Ridership weighted'] = scores['Average Ridership'] * weights[2]
scores['Total Weighted Ranking'] = scores['Density weighted'] + scores['Usage weighted'] + scores['Ridership weighted']

scores = scores.sort_values(by=['Total Weighted Ranking'])
scores
```

```python
scores.drop(columns=['Total Weighted Ranking','Bus Stop Density','Bus Stop Usage', 'Average Ridership']).plot(x='Neighborhood', 
                                                                                                              kind="bar", 
                                                                                                              stacked=True,
                                                                                                              figsize=(15,10)
                                                                                                             )
```

Sweeeeeet. Stacked bar charts work great here because we can obtain all the information we need at a glance to make an informed decision about which neighborhood deserves the title of the **best bus stop neighborhood in Pittsburgh**. However, the above visualization is so cluttered particularly because we have so many neighborhoods. Let's *trim* it down to only show the top few neighborhoods in contest for the top spot.

```python
plt.style.use('seaborn-pastel')
scores.head(10).drop(columns=['Total Weighted Ranking','Density weighted','Usage weighted','Ridership weighted']).plot(x='Neighborhood', 
                                                                                                                       kind="bar",
                                                                                                                       stacked=True,
                                                                                                                       figsize=(10,6)
                                                                                                                      )
```

Note that this plot is of *'Bus Stop Density'*, *'Bus Stop Usage'* and *'Average Ridership'* rankings **before being weighted**! Still, it is clear that **Friendship**, **North Oakland**, **East Liberty** and **Shadyside** stand out here.

Let's take a look at the actual weighted data and rankings.

```python
plt.style.use('dark_background')
scores.head(10).drop(columns=['Total Weighted Ranking','Bus Stop Density','Bus Stop Usage', 'Average Ridership']).plot(x='Neighborhood', 
                                                                                                                       kind="bar",
                                                                                                                       stacked=True,
                                                                                                                       figsize=(10,6)
                                                                                                                      )
```

Alas, the overall **best bus stop neighborhoods** in Pittsburgh are:

```python
scores['Overall'] = 1 / scores['Total Weighted Ranking']

scores = scores.merge(hoods_TSusage[['Neighborhood','geometry']], how="outer", on="Neighborhood")
scores = gpd.GeoDataFrame(scores, geometry='geometry')
```

```python
scores.plot(column='Overall', 
            cmap='PuBuGn', 
            edgecolor='lightblue', 
            legend=True,
            legend_kwds={'label':'Overall Bus Stop Ranking'},
            figsize=(12,8)
           )# plot the geopandas dataframe
```
